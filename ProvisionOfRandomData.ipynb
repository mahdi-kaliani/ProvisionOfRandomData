```python
import numpy as np
import matplotlib.pyplot as plt


num_data = 300
np.random.seed(0)
x_data = np.random.uniform(low=-1, high=1, size=(num_data, 2))
y_data = np.zeros(shape=(num_data,))

for i in range(num_data):
    if x_data[i, 0] + x_data[i, 1] > 0.5:
        y_data[i] = 1
        x_data[i, :] += np.array([0.5, 0.5])
    elif x_data[i, 0] - x_data[i, 1] > 0.5:
        y_data[i] = 2
        x_data[i, :] += np.array([0.5, -0.5])
    else:
        y_data[i] = 3
        x_data[i, :] += np.array([-0.5, 0.5])

plt.scatter(x_data[y_data == 1, 0], x_data[y_data == 1, 1], c='blue', label='Class 1')
plt.scatter(x_data[y_data == 2, 0], x_data[y_data == 2, 1], c='red', label='Class 2')
plt.scatter(x_data[y_data == 3, 0], x_data[y_data == 3, 1], c='green', label='Class 3')
plt.legend()
plt.show()
```
